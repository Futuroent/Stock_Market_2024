## Apply Feature Scaling


```python
# Import necessary libraries
from sklearn.decomposition import PCA
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
import joblib

# Load your data (ensure you have 4 features in X)
X = apple_data[['Daily Return', '20-Day MA', '50-Day MA', 'Volume']]  # Features
y = apple_data['Adj Close']  # Target variable

# Step 1: Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components (you can choose between 1 to 4)
X_pca = pca.fit_transform(X_scaled)

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Step 4: Initialize the Ridge regression model
ridge_model = Ridge(alpha=1.0)

# Step 5: Train the model
ridge_model.fit(X_train, y_train)

# Step 6: Cross-Validation with PCA features
ridge_scores_pca = cross_val_score(ridge_model, X_pca, y, cv=5, scoring='neg_mean_squared_error')
ridge_mse_scores_pca = -ridge_scores_pca
print(f"Cross-Validation MSE with Ridge after PCA: {ridge_mse_scores_pca.mean()} ± {ridge_mse_scores_pca.std()}")

# Step 7: Save the trained model and scaler
joblib.dump(ridge_model, '../models/ridge_model_final.pkl')
joblib.dump(scaler, '../models/standard_scaler.pkl')
joblib.dump(pca, '../models/pca_transform.pkl')  # Save the PCA model as well

print("Model, scaler, and PCA transformer saved successfully.")

# Step 8: Make predictions on the test set
y_pred = ridge_model.predict(X_test)

# Step 9: Evaluate the model
mse_test = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error on Test Set: {mse_test}")
```

## Fine-tune the Linear Regression model Ridge or Lasso (Optional)

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error
import joblib

# Assume that X_scaled and y are already prepared and scaled, and PCA has been applied if needed.

# Split the data into training and testing sets (reusing X_pca and y)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# ----------------------- Ridge Regression Fine-Tuning -----------------------

# Define the Ridge model
ridge_model = Ridge()

# Create a dictionary of hyperparameters to tune
ridge_params = {
    'alpha': [0.01, 0.1, 1.0, 10.0, 100.0]
}

# Use GridSearchCV for hyperparameter tuning with cross-validation
ridge_grid = GridSearchCV(ridge_model, ridge_params, cv=5, scoring='neg_mean_squared_error')

# Train the model
ridge_grid.fit(X_train, y_train)

# Best hyperparameters and best score
best_ridge_model = ridge_grid.best_estimator_
ridge_best_alpha = ridge_grid.best_params_['alpha']
ridge_best_score = -ridge_grid.best_score_

print(f"Best Ridge Alpha: {ridge_best_alpha}")
print(f"Best Ridge CV MSE: {ridge_best_score}")

# Save the best Ridge model
joblib.dump(best_ridge_model, '../models/best_ridge_model.pkl')

# ----------------------- Lasso Regression Fine-Tuning -----------------------

# Define the Lasso model
lasso_model = Lasso()

# Create a dictionary of hyperparameters to tune
lasso_params = {
    'alpha': [0.01, 0.1, 1.0, 10.0, 100.0]
}

# Use GridSearchCV for hyperparameter tuning with cross-validation
lasso_grid = GridSearchCV(lasso_model, lasso_params, cv=5, scoring='neg_mean_squared_error')

# Train the model
lasso_grid.fit(X_train, y_train)

# Best hyperparameters and best score
best_lasso_model = lasso_grid.best_estimator_
lasso_best_alpha = lasso_grid.best_params_['alpha']
lasso_best_score = -lasso_grid.best_score_

print(f"Best Lasso Alpha: {lasso_best_alpha}")
print(f"Best Lasso CV MSE: {lasso_best_score}")

# Save the best Lasso model
joblib.dump(best_lasso_model, '../models/best_lasso_model.pkl')

# ----------------------- Evaluate on the Test Set -----------------------

# Evaluate Ridge on the test set
y_pred_ridge = best_ridge_model.predict(X_test)
ridge_test_mse = mean_squared_error(y_test, y_pred_ridge)
print(f"Ridge Test Set MSE: {ridge_test_mse}")

# Evaluate Lasso on the test set
y_pred_lasso = best_lasso_model.predict(X_test)
lasso_test_mse = mean_squared_error(y_test, y_pred_lasso)
print(f"Lasso Test Set MSE: {lasso_test_mse}")

# ----------------------- Save Results -----------------------

# Save the Ridge and Lasso models and scalers
joblib.dump(best_ridge_model, '../models/ridge_model_final.pkl')
joblib.dump(best_lasso_model, '../models/lasso_model_final.pkl')
joblib.dump(scaler, '../models/standard_scaler.pkl')
joblib.dump(pca, '../models/pca_transform.pkl')  # If PCA was applied

```

#### Implement Random Forest Regressor

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import joblib

# Step 1: Ensure X_scaled and y_filled_cleaned are of the same length
# Assume X_scaled and y_filled_cleaned are your feature matrix and target variable

# Convert X_scaled to DataFrame (if it's not already)
X_scaled_df = pd.DataFrame(X_scaled)

# Convert y_filled_cleaned to pandas Series (if it's not already)
y_filled_cleaned_series = pd.Series(y_filled_cleaned)

# Combine the features and target into a single DataFrame
combined_df = pd.concat([X_scaled_df, y_filled_cleaned_series], axis=1)

# Drop rows with any missing values
combined_cleaned = combined_df.dropna()

# Separate the features and target again
X_scaled_cleaned = combined_cleaned.iloc[:, :-1].values  # All columns except the last one (X_scaled)
y_filled_cleaned = combined_cleaned.iloc[:, -1].values   # The last column (y_filled_cleaned)

# Check shapes again after cleaning to ensure they match
print(f"X_scaled_cleaned shape: {X_scaled_cleaned.shape}")
print(f"y_filled_cleaned shape: {y_filled_cleaned.shape}")

# Step 2: Split the data into training and testing sets
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled_cleaned, y_filled_cleaned, test_size=0.2, random_state=42)

# Step 3: Create a Random Forest Regressor model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Step 4: Perform cross-validation to evaluate the model
rf_scores = cross_val_score(rf_model, X_scaled_cleaned, y_filled_cleaned, cv=5, scoring='neg_mean_squared_error')

# Convert negative MSE to positive values
rf_mse_scores = -rf_scores

# Output the Cross-Validation MSE and standard deviation
print(f"Cross-Validation MSE with Random Forest: {rf_mse_scores.mean()} ± {rf_mse_scores.std()}")

# Step 5: Train the Random Forest model on the training set
rf_model.fit(X_train_scaled, y_train)

# Step 6: Make predictions using the Random Forest model on the test set
y_pred_rf = rf_model.predict(X_test_scaled)

# Step 7: Evaluate the model's performance on the test set
mse_rf = mean_squared_error(y_test, y_pred_rf)
print(f"Mean Squared Error with Random Forest on Test Set: {mse_rf}")

# Step 8: Save the trained Random Forest model and scaler
joblib.dump(rf_model, '../models/random_forest_model_cleaned.pkl')  # Save the trained Random Forest model
joblib.dump(scaler, '../models/scaler.pkl')  # Save the scaler used for scaling
```

## Explore an alternative model 

####  1-Implement ARIMA Model

```python
from statsmodels.tsa.stattools import adfuller

# Function to check if data is stationary
def check_stationarity(timeseries):
    result = adfuller(timeseries)
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    if result[1] > 0.05:
        print("The data is non-stationary.")
    else:
        print("The data is stationary.")

# Apply on the 'Adj Close' column
check_stationarity(apple_data['Adj Close'])
```

#### 1-Fix the Differencing Process

```python
# Perform differencing to remove trend, without dropping NaNs too early
apple_data['Adj Close Differenced'] = apple_data['Adj Close'].diff()

# Now drop NaN values created by differencing
apple_data = apple_data.dropna(subset=['Adj Close Differenced'])

# Check stationarity again after fixing the differencing process
check_stationarity(apple_data['Adj Close Differenced'])
```

#### 2-Ensure the Data is Clean

```python
# Drop rows with NaN values
apple_data = apple_data.dropna()

# Now try the stationarity check again
check_stationarity(apple_data['Adj Close Differenced'])
```

#### 3-Proceed with ARIMA After Differencing

##### 1-1 Ensure Proper Date Index

```python
# Convert the 'Date' column to datetime if not already done
apple_data.index = pd.to_datetime(apple_data.index)

# Set frequency to the date index if it's daily data
apple_data = apple_data.asfreq('B')  # 'B' stands for business days

# Verify that the frequency is set
print(apple_data.index.freq)
```

##### 1-2 Refit the ARIMA Model

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model on the differenced data
arima_model = ARIMA(apple_data['Adj Close Differenced'], order=(5, 1, 0))  # ARIMA(5,1,0) as an example
arima_result = arima_model.fit()
forecast = arima_result.forecast(steps=30)


# Print summary of the ARIMA model
print(arima_result.summary())
```

#### 4-Forecasting and Model Evaluation

##### 4-1 Check for NaN values in the actual and forecasted data

```python
print(apple_data['Adj Close'].iloc[-30:].isna().sum())
print(forecast.isna().sum())
```

##### 4-2 Check Data Lengths Before and After Removing NaN Values


```python
# Check lengths before removing NaN
print(f"Original length of actual data: {len(apple_data['Adj Close'].iloc[-30:])}")
print(f"Original length of forecast: {len(forecast)}")

# Check lengths after removing NaN
print(f"Length of actual data after dropna: {len(actual_data)}")
print(f"Length of forecast after dropna: {len(forecast)}")
```

##### 4-3 Fill NaN values in the actual data

```python
actual_data = apple_data['Adj Close'].iloc[-30:].ffill()  # Use ffill directly instead of fillna(method='ffill')
forecast = forecast.dropna() 

print(f"Length of actual data after filling NaNs: {len(actual_data)}")
```

##### 4-4 Ensure Matching Data Lengths

```python
# Ensure both have the same length after handling NaNs
if len(actual_data) == len(forecast):
    # Compute MSE
    mse_arima = mean_squared_error(actual_data, forecast)
    print(f"Mean Squared Error with ARIMA: {mse_arima}")
else:
    print("Actual and forecasted data lengths still do not match.")
```
